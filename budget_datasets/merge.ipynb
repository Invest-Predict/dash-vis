```python
def merge_csv_files(march_file, april_file, may_file, output_file):
    # Загружаем данные
    df_march = pd.read_csv(march_file, parse_dates=['datetime'])
    df_april = pd.read_csv(april_file, parse_dates=['datetime'])
    df_may = pd.read_csv(may_file, parse_dates=['datetime'])
    
    # Добавляем колонку с приоритетом
    df_march['priority'] = 1
    df_april['priority'] = 2
    df_may['priority'] = 3
    
    # Объединяем
    df = pd.concat([df_march, df_april, df_may])
    
    # Оставляем только последние версии записей
    df = df.sort_values(by=['datetime', 'priority'], ascending=[True, False])
    df = df.drop_duplicates(subset=['datetime'], keep='first').drop(columns=['priority'])
    
    # Сохраняем результат
    df.to_csv(output_file, index=False)
    print(f"Объединенный файл сохранен как {output_file}")
```

```python
import pandas as pd

def merge_csv_files(march_file, april_file, may_file, output_file):
    # Загрузка данных с проверкой формата дат
    df_march = pd.read_csv(march_file, parse_dates=['datetime'])
    df_april = pd.read_csv(april_file, parse_dates=['datetime'])
    df_may = pd.read_csv(may_file, parse_dates=['datetime'])
    
    # Обрезка по последнему дню каждого месяца
    df_march = df_march[df_march['datetime'].dt.month == 3]
    df_april = df_april[df_april['datetime'].dt.month == 4]
    df_may = df_may[df_may['datetime'].dt.month == 5]
    
    # Объединение
    merged_df = pd.concat([df_march, df_april, df_may])
    merged_df.to_csv(output_file, index=False)
    print("Данные успешно объединены!")
```

```python
import pandas as pd

def merge_csv_files(first_file, second_file, output_file):
    # Загрузка данных с проверкой формата дат
    df_first = pd.read_csv(first_file, parse_dates=['datetime'])
    df_second = pd.read_csv(second_file, parse_dates=['datetime'])
    
    # Объединение
    merged_df = pd.concat([df_first, df_second])
    merged_df.to_csv(output_file, index=False)
    print("Данные успешно объединены!")
```

```python
import pandas as pd

def color_csv_files(file, output_file):
    # Загрузка данных с проверкой формата дат
    df = pd.read_csv(file, parse_dates=['datetime'])
    df["color"] = "gray"
    df.loc[df["budget"].diff(-1) > 0, "color"] = "red"
    df.loc[df["budget"].diff(-1) < 0, "color"] = "green"
    
    
    df.to_csv(output_file, index=False)
    print("Данные успешно покрашены!")
```

```python
color_csv_files('Gazprom38.csv', 'Gazprom38.csv')
color_csv_files('MMK38.csv', 'MMK38.csv')
color_csv_files('MTS38.csv', 'MTS38.csv')
color_csv_files('Tatneft38.csv', 'Tatneft38.csv')
color_csv_files('Whoosh38.csv', 'Whoosh38.csv')
color_csv_files('VTB38.csv', 'VTB38.csv')
```

```python
import dash
from dash import dcc, html, Output, Input
import pandas as pd
import plotly.express as px
import os

# Загружаем все файлы, оканчивающиеся на '35.csv'
files = [f for f in os.listdir() if f.endswith("35.csv")]
data = {}

for file in files:
    company_name = file.replace("35.csv", "")
    df = pd.read_csv(file, parse_dates=["datetime"])
    data[company_name] = df

# Загружаем файлы со статистикой
stat_files = [f for f in os.listdir() if f.endswith("_1_min.csv")]
stats_data = {}

for file in stat_files:
    company_name = file.replace("_stats.csv", "")
    df = pd.read_csv(file, parse_dates=["utc"])
    stats_data[company_name] = df

# Создаем Dash-приложение
app = dash.Dash(__name__)

all_dates = pd.concat([df["datetime"] for df in data.values()])
min_date, max_date = all_dates.min(), all_dates.max()

# Функция для генерации меток слайдера
def generate_marks(start, end):
    dates = pd.date_range(start=start, end=end, freq='5D')
    month_starts = pd.date_range(start=start, end=end, freq='MS')
    marks = {int(date.timestamp()): date.strftime('%Y-%m-%d') for date in dates}
    marks.update({int(date.timestamp()): date.strftime('%Y-%m-%d') for date in month_starts})
    return marks

app.layout = html.Div([
    dcc.Checklist(
        id='company-selector',
        options=[{'label': name, 'value': name} for name in data.keys()],
        value=list(data.keys()),
        inline=True
    ),
    dcc.RangeSlider(
        id='date-slider-budget',
        min=min_date.timestamp(),
        max=max_date.timestamp(),
        value=[min_date.timestamp(), max_date.timestamp()],
        marks=generate_marks(min_date, max_date),
        step=432000  # 5 дней в секундах
    ),
    dcc.Graph(id='budget-graph'),
    dcc.Dropdown(
        id='single-company-selector',
        options=[{'label': name, 'value': name} for name in data.keys()],
        value=list(data.keys())[0],
        clearable=False
    ),
    dcc.Checklist(
        id='strategy-selector',
        options=[],
        value=[],
        inline=True
    ),
    dcc.RangeSlider(
        id='date-slider-strategy',
        min=min_date.timestamp(),
        max=max_date.timestamp(),
        value=[min_date.timestamp(), max_date.timestamp()],
        marks=generate_marks(min_date, max_date),
        step=432000  # 5 дней в секундах
    ),
    dcc.Graph(id='strategy-graph'),
    dcc.Dropdown(
        id='stats-company-selector',
        options=[{'label': name, 'value': name} for name in stats_data.keys()],
        value=list(stats_data.keys())[0],
        clearable=False
    ),
    dcc.Checklist(
        id='stats-metric-selector',
        options=[
            {'label': 'Budget', 'value': 'budget'},
            {'label': 'Close', 'value': 'close'},
            {'label': 'Volume', 'value': 'volume'}
        ],
        value=['budget', 'close'],
        inline=True
    ),
    dcc.RangeSlider(
        id='date-slider-stats',
        min=min_date.timestamp(),
        max=max_date.timestamp(),
        value=[min_date.timestamp(), max_date.timestamp()],
        marks=generate_marks(min_date, max_date),
        step=432000  # 5 дней в секундах
    ),
    dcc.Graph(id='stats-graph')
])

@app.callback(
    Output('budget-graph', 'figure'),
    Input('company-selector', 'value'),
    Input('date-slider-budget', 'value')
)
def update_budget_graph(selected_companies, date_range):
    start_date = pd.to_datetime(date_range[0], unit='s')
    end_date = pd.to_datetime(date_range[1], unit='s')
    
    filtered_data = []
    for company in selected_companies:
        df = data[company]
        df_filtered = df[(df["datetime"] >= start_date) & (df["datetime"] <= end_date)]
        df_filtered["company"] = company
        filtered_data.append(df_filtered)
    
    if not filtered_data:
        return px.line(title="Нет данных для отображения")
    
    df_final = pd.concat(filtered_data)
    fig = px.line(df_final, x="datetime", y="budget", color="company", title="Бюджет по времени")
    return fig

@app.callback(
    Output('strategy-selector', 'options'),
    Input('single-company-selector', 'value')
)
def update_strategy_options(selected_company):
    strategies = data[selected_company]['strategy'].unique()
    return [{'label': strat, 'value': strat} for strat in strategies]

@app.callback(
    Output('strategy-graph', 'figure'),
    Input('single-company-selector', 'value'),
    Input('strategy-selector', 'value'),
    Input('date-slider-strategy', 'value')
)
def update_strategy_graph(selected_company, selected_strategies, date_range):
    start_date = pd.to_datetime(date_range[0], unit='s')
    end_date = pd.to_datetime(date_range[1], unit='s')
    
    df = data[selected_company]
    df_filtered = df[(df["datetime"] >= start_date) & (df["datetime"] <= end_date) & (df["strategy"].isin(selected_strategies))]
    df_filtered["color"] = "gray"
    df_filtered.loc[df_filtered["budget"].diff() > 0, "color"] = "green"
    df_filtered.loc[df_filtered["budget"].diff() < 0, "color"] = "red"
    
    if df_filtered.empty:
        return px.scatter(title="Нет данных для отображения")
    
    fig = px.scatter(df_filtered, x="datetime", y="budget", color="color", title=f"Бюджет по стратегиям для {selected_company}", color_discrete_map={"red": "red", "green": "green", "gray": "gray"})
    return fig

@app.callback(
    Output('stats-graph', 'figure'),
    Input('stats-company-selector', 'value'),
    Input('stats-metric-selector', 'value'),
    Input('date-slider-stats', 'value')
)
def update_stats_graph(selected_company, metrics, date_range):
    start_date = pd.to_datetime(date_range[0], unit='s')
    end_date = pd.to_datetime(date_range[1], unit='s')
    
    # Объединяем данные из двух источников
    df_stats = stats_data[selected_company]
    df_budget = data[selected_company[:-10]]
    
    # Фильтруем по дате
    df_stats = df_stats[(df_stats["utc"] >= start_date) & (df_stats["utc"] <= end_date)]
    df_budget = df_budget[(df_budget["datetime"] >= start_date) & (df_budget["datetime"] <= end_date)]
    
    # Объединяем по дате
    merged_df = pd.merge(
        df_budget[['datetime', 'budget']],
        df_stats[['utc', 'close', 'volume']],
        left_on='datetime',
        right_on='utc',
        how='inner'
    )
    
    if merged_df.empty:
        return px.line(title="Нет данных для отображения")
    
    # Создаем фигуру с вторичной осью
    fig = px.line(merged_df, x='datetime', y=metrics[0])
    
    # Добавляем вторую метрику на вторичную ось
    if len(metrics) > 1:
        fig.add_scatter(
            x=merged_df['datetime'],
            y=merged_df[metrics[1]],
            name=metrics[1],
            yaxis='y2'
        )
        
        # Настраиваем оси
        fig.update_layout(
            yaxis2=dict(
                title=metrics[1],
                overlaying='y',
                side='right'
            )
        )
    
    fig.update_layout(title=f"{', '.join(metrics)} для {selected_company}")
    return fig


if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
from dash import dcc, html, Output, Input
import pandas as pd
import plotly.express as px
import os

# Загружаем все файлы, оканчивающиеся на '35.csv'
files = [f for f in os.listdir() if f.endswith("35.csv")]
data = {}

for file in files:
    company_name = file.replace("35.csv", "")
    df = pd.read_csv(file, parse_dates=["datetime"])
    data[company_name] = df

# Создаем Dash-приложение
app = dash.Dash(__name__)

all_dates = pd.concat([df["datetime"] for df in data.values()])
min_date, max_date = all_dates.min(), all_dates.max()

def generate_marks(start, end, num=10):
    dates = pd.date_range(start=start, end=end, periods=num)
    return {int(date.timestamp()): date.strftime('%Y-%m-%d') for date in dates}

app.layout = html.Div([
    dcc.Checklist(
        id='company-selector',
        options=[{'label': name, 'value': name} for name in data.keys()],
        value=list(data.keys()),
        inline=True
    ),
    dcc.RangeSlider(
        id='date-slider',
        min=min_date.timestamp(),
        max=max_date.timestamp(),
        value=[min_date.timestamp(), max_date.timestamp()],
        marks=generate_marks(min_date, max_date),
        step=None
    ),
    dcc.Graph(id='budget-graph')
])

@app.callback(
    Output('budget-graph', 'figure'),
    Input('company-selector', 'value'),
    Input('date-slider', 'value')
)
def update_graph(selected_companies, date_range):
    start_date = pd.to_datetime(date_range[0], unit='s')
    end_date = pd.to_datetime(date_range[1], unit='s')
    
    filtered_data = []
    for company in selected_companies:
        df = data[company]
        df_filtered = df[(df["datetime"] >= start_date) & (df["datetime"] <= end_date)]
        df_filtered["company"] = company
        filtered_data.append(df_filtered)
    
    if not filtered_data:
        return px.line(title="Нет данных для отображения")
    
    df_final = pd.concat(filtered_data)
    fig = px.line(df_final, x="datetime", y="budget", color="company", title="Бюджет по времени")
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
